```python
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import math
```

```python
data = pd.read_csv("../../Z_Resources/Statistics with Python/Inferential Statistics/Capstone Project - Lights, Camera, Statistics!/imdb.csv")
data.head(1)
```

```python
df = data[["IMDB_Rating","Series_Title", "Released_Year","Certificate", "Runtime","Genre", "Meta_score", "Director", "Gross","No_of_Votes"]]
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df["Released_Year"] = pd.to_numeric(df["Released_Year"],errors = "coerce")
```

```python
df["Runtime"] = df["Runtime"].map(lambda n: int(n.replace(" min","")))
```

```python
def gross_fun(s):
    try:
        return float("".join(s.split(",")))
    except:
        return s
```

```python
df["Gross"] =  df["Gross"].map(gross_fun)
```
