```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("../Data/Churn_Modelling.csv")
```

```python
data.head()
```

```python
telco_new = data[["Geography","Gender","Exited"]]
```

```python
telco_new.head(5)
```

```python
for i, pred in enumerate(telco_new.drop(columns=["Exited"])):
    plt.figure()
    sns.countplot(x=pred,data=data,hue="Exited")
```

### Bivariate

```python
sns.histplot(x="Gender",hue="Geography",stat= "count",data=data,multiple="dodge")
```

```python
telco_exited = telco_new[telco_new["Exited"]==1]
```

```python
sns.histplot(x="Gender",hue="Geography",stat="count",data=telco_exited,multiple="dodge")
```

```python
g=sns.histplot(x="Gender",hue="Geography",stat= "count",data=data,multiple="dodge" ,palette="magma")
sns.histplot(x="Gender",hue="Geography",stat="count",data=telco_exited,multiple="dodge")
```

```python
data.corr(numeric_only = True)
```

```python
data.corr(numeric_only=True)["Exited"].sort_values(ascending=False).plot(kind="bar")
```

```python
sns.heatmap(data=data.corr(numeric_only = True),cmap="Paired")
```

```python
data["Age"].value_counts().sort_index().plot(kind="line")
```

```python
data[data["Exited"]==1]["Age"].value_counts().sort_index().plot()
```

```python
Tot = sns.kdeplot(data.Age[(data["Exited"] == 0)], color="Red", fill=True)
Tot = sns.kdeplot(data.Age[(data["Exited"] == 1)], color="Blue", fill=True)
Tot.legend(["No Churn","Churn"], loc="upper right")
Tot.set_title("Age by Chrun")
```
