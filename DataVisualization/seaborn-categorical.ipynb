```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset("tips")
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x="sex",y="tip",data=tips,estimator=np.var)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(x="sex",y="total_bill", data=tips, hue="smoker")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="smoker")
```

```python
sns.stripplot(x="day",y="total_bill", data=tips, hue="sex")
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="violin", hue="smoker", split=True)
```
