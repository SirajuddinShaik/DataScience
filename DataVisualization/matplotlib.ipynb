```python
books={
    "dannie":90,
    "brouke":1000,
    "djalok":100
}
fig,ax =plt.subplots(figsize=(5,5))
ax.set(title="books",
      xlabel="book_name",
      ylabel="price")
ax.bar(books.keys(),height=books.values());
```

```python
import pandas as pd
df = pd.read_csv("zero-to-mastery-ml/data/car-sales.csv")
df["Price"]=df["Price"].str.replace('$','').str.replace(',','').str.replace('.','').str[:-2]
df["date"] = pd.date_range("01/06/2003",periods=10)
```

```python
df
```

```python
df["Total Sales"] = df["Price"].astype("int").cumsum()
```

```python
df
```

```python
df.plot(x="date",y="Total Sales")
```
